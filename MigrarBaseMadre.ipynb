```python
# ==============================================================================
# PROYECTO COLAB - CONTROL DE DESEMPEÑO ESTUDIANTIL HE
# Objetivo: Sobrescribir Base 2 (Hoja: BASE) con la información de Base 1 (Hoja: GENERAL)
#           Importando desde la fila 2 (Encabezados inclusive).
# ==============================================================================

# 1. Autenticación con Google Drive y Sheets en Google Colab
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)

# 2. Configuración de enlaces de Sheets
# Identificadores extraídos directamente de los enlaces provistos por ti
ID_BASE_1 = "1FkxpE-OnT05R6rDXI5lZTE1gaJaMIenaekHt3tHooRE" # Data Notas / Origen
ID_BASE_2 = "15PJ0iRbpSmmYTpsvMDCRs6RYI6onx595VLcZtl9f2BY" # Seguimiento / Destino

PESTANA_ORIGEN = "GENERAL"
PESTANA_DESTINO = "BASE"

try:
    print("🔄 Conectando con las Bases de Datos de High Education...")
    # Abrir las hojas usando los IDs seguros de Google
    doc_origen = gc.open_by_key(ID_BASE_1)
    doc_destino = gc.open_by_key(ID_BASE_2)

    hoja_origen = doc_origen.worksheet(PESTANA_ORIGEN)
    hoja_destino = doc_destino.worksheet(PESTANA_DESTINO)

    # 3. Leer los datos desde la FILA 2 (que contiene los encabezados en la Base 1)
    # Obtenemos todos los valores de la pestaña de origen
    todos_los_datos = hoja_origen.get_all_values()

    # Validamos que haya información suficiente
    if len(todos_los_datos) < 2:
        print("⚠️ Advertencia: La hoja de origen tiene menos de 2 filas de información. Proceso cancelado.")
    else:
        # Filtramos para conservar desde el índice 1 en adelante (Fila 2 en formato de usuario)
        datos_a_copiar = todos_los_datos[1:]

        # 4. Sobrescribir Base 2
        # Limpiamos todo el contenido anterior de la pestaña destino para asegurar una base limpia
        hoja_destino.clear()
        print(f"🧹 Pestaña '{PESTANA_DESTINO}' vaciada de registros antiguos.")

        # Insertamos la información a partir de la celda A1 en la pestaña de destino
        hoja_destino.update('A1', datos_a_copiar)
        print(f"✅ ¡Transferencia Exitosa! Se copiaron {len(datos_a_copiar)} filas (incluyendo la fila de encabezados) a la pestaña '{PESTANA_DESTINO}'.")

except Exception as e:
    print(f"❌ Error en el proceso: {e}")
    print("Verifica tener permisos de editor en ambas hojas con la cuenta de Google utilizada en Colab.")
```
